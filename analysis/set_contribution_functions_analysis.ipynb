# --- **Gradient Set Contribution Function** ---

## Semantic: **DFQuAD**

```python
from qbaf_ctrbs.gradient import determine_gradient_ctrb

args = ['a', 'b', 'c']
initial_strengths = [0.5, 1, 1]
atts = [('b', 'a'), ('c', 'a')]
supps = []
qbaf = QBAFramework(args, initial_strengths, atts, supps, semantics="DFQuAD_model")

ctrb_b_to_a = determine_gradient_ctrb('a', {'b'}, qbaf)
ctrb_c_to_a = determine_gradient_ctrb('a', {'c'}, qbaf)
ctrb_bc_to_a = determine_gradient_ctrb('a', {'b', 'c'}, qbaf)

print('Contribution b to a, gradient:', ctrb_b_to_a)
print('Contribution c to a, gradient:', ctrb_c_to_a)
print('Contribution {b, c} to a, gradient:', ctrb_bc_to_a)

visualize(qbaf, with_fs=True, round_to=6)
```

# --- **Helper Functions** ---

```python
##############################################################################
# HELPER FUNCTIONS
#
# These functions creates a visual representation
# of the contribution of one or more arguments.
##############################################################################
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.patches import Patch
from collections.abc import Callable

from qbaf_ctrbs.shapley import determine_shapley_ctrb


def build_qbaf_with_d_f_strengths(qbaf, new_strength_d, new_strength_f, arg_d='d', arg_f='f'):
    """Return a copy of qbaf where d and f have the provided initial strengths."""
    arguments = list(qbaf.arguments)
    updated_initial_strengths = []
    for arg in arguments:
        if arg == arg_d:
            updated_initial_strengths.append(new_strength_d)
        elif arg == arg_f:
            updated_initial_strengths.append(new_strength_f)
        else:
            updated_initial_strengths.append(qbaf.initial_strength(arg))
    return QBAFramework(
        arguments,
        updated_initial_strengths,
        qbaf.attack_relations.relations,
        qbaf.support_relations.relations,
        semantics=qbaf.semantics,
    )

def compute_signs(
    base_qbaf,
    contribution_func: Callable,
    strength_d,
    strength_f,
    topic='a',
    arg_d='d',
    arg_f='f',
    epsilon=1e-9,
):
    """Compute SIGN (-1/0/+1) of contributions of {d}, {f}, and {d,f} to the topic."""
    qbaf_at_point = build_qbaf_with_d_f_strengths(base_qbaf, strength_d, strength_f, arg_d, arg_f)
    contrib_d  = contribution_func(topic, {arg_d}, qbaf_at_point)
    contrib_f  = contribution_func(topic, {arg_f}, qbaf_at_point)
    contrib_df = contribution_func(topic, {arg_d, arg_f}, qbaf_at_point)

    def sign_of_value(value):
        if value < -epsilon: return -1
        if value >  epsilon: return +1
        return 0

    return sign_of_value(contrib_d), sign_of_value(contrib_f), sign_of_value(contrib_df)

def plot_sign_map(
    base_qbaf,
    contribution_func: Callable,
    use_marker:bool = True,
    topic='a',
    arg_d='d',
    arg_f='f',
    resolution=100,
    epsilon=1e-9,
):
    """2D map over initial strengths of {d,f}. GREEN if sign(d) == sign(f) != sign({d,f}); GRAY otherwise."""
    grid_d_values = np.linspace(0.0, 1.0, resolution)
    grid_f_values = np.linspace(0.0, 1.0, resolution)
    class_code_grid = np.zeros((resolution, resolution), dtype=int)

    for col_index, strength_d in enumerate(grid_d_values):
        for row_index, strength_f in enumerate(grid_f_values):
            sign_d, sign_f, sign_df = compute_signs(
                base_qbaf,
                contribution_func,
                strength_d,
                strength_f,
                topic=topic,
                arg_d=arg_d,
                arg_f=arg_f,
                epsilon=epsilon,
            )
            if (sign_d == sign_f) and (sign_df != sign_d):
                class_code_grid[row_index, col_index] = 1
            else:
                class_code_grid[row_index, col_index] = 0

    # color_map = ListedColormap(['#dddddd', '#2ecc71'])  # gray, green
    color_map = ListedColormap(['#dddddd', '#555555'])  # light gray, dark gray

    plt.figure(figsize=(6, 5))
    plt.imshow(
        class_code_grid, origin='lower', extent=[0, 1, 0, 1],
        cmap=color_map, interpolation='nearest', aspect='auto'
    )

    if use_marker:
      baseline_d = base_qbaf.initial_strength(arg_d)
      baseline_f = base_qbaf.initial_strength(arg_f)
      plt.scatter([baseline_d], [baseline_f], marker='x', s=80, linewidths=2, color='k')

    plt.xlabel(f'Initial strength of {arg_d}')
    plt.ylabel(f'Initial strength of {arg_f}')
    plt.title(f'Sign map for contribution of {{{arg_d}}}, {{{arg_f}}}, and {{{arg_d}, {arg_f}}} to topic {topic}')

    legend_items = [
        # Patch(facecolor='#2ecc71', label=f'sign({{{arg_d}}}) = sign({{{arg_f}}}) ≠ sign({{{arg_d},{arg_f}}})'),
        Patch(facecolor='#555555', label=f'sign({{{arg_d}}}) = sign({{{arg_f}}}) ≠ sign({{{arg_d},{arg_f}}})'),
        Patch(facecolor='#dddddd', label='otherwise'),
    ]
    plt.legend(handles=legend_items, loc='upper left')
    plt.show()
```
